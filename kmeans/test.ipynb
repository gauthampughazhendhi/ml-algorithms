```python
import numpy as np
import matplotlib.pyplot as plt
from kmeans import KMeans
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs


np.random.seed(42)
```

```python
X, y = make_blobs(
    centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40
)
print("Input shape:", X.shape)
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1234)

clusters = len(np.unique(y))
print("k:", clusters)

k = KMeans(k=clusters, n_iter=300)
k.fit(train_X)

predictions = k.predict(train_X)
k.plot(train_X, predictions)
```

```python
predictions = k.predict(test_X)
k.plot(test_X, predictions)
```
